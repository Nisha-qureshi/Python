```python
x=np.random.randint(0,50,30)
y=np.random.random(30)
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.1,0.1,0.4,0.4])
ax3=fig.add_axes([0.6,0.1,0.34,0.4])
ax4=fig.add_axes([0.1,0.6,0.8,0.36])
ax2.plot(x,label="x")
ax2.legend() #show the labels into figure
ax3.plot(y,"ro-",label="y")
ax3.legend()
ax4.plot(x,label="x")
ax4.plot(x**3,"k:",label="$ x^3 $")
ax4.legend()
plt.show()
```
