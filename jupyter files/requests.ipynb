```python
import requests
import matplotlib.pyplot as plt   
for i in range(len(url_list)):
    fname="myimg"+str(i)+".jpg"
    img=requests.get(url_list[i])
    if img.status_code==200 and b'\\x' in img.content:
        with open(fname,"wb") as fp:
            fp.write(img.content)
            fp.close()
            dev=plt.imread(fname)
            plt.imshow(dev)
            plt.xticks([])
            plt.yticks([])
            plt.show()
```
