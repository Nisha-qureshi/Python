```python
fig=plt.figure()
f1=fig.add_axes([0,0,1,1])
ax1=fig.add_axes([0.5,0.3,0.4,0.5])
y1=x
y2=x**2
y3=x**3
f1.plot(y1,label="x",c="r")  # if we don't provide label then legend will be a empty box
f1.legend()
ax1.plot(y2,"g--",y3,label="x^2")
ax1.legend()
```

```python
from matplotlib import pyplot as plt
plt.plot([1,9,3],[4,5,6])
plt.title("first graph",fontsize=20)
plt.xlabel("x -->",color="r",fontsize=16)
plt.ylabel("y -->",color="r",fontsize=16)
plt.grid()    #---> will show grid on graph
plt.show()
#%matplot inline -->to show line to line
```

## linear graph

```python

from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot") # shows light blue color in background
plt.plot([1,9,3],"g",label="[1,9,3]",linewidth=2)
plt.plot([4,5,6],c="c",label="[4,5,6]",lw=4)
plt.title("first graph",fontsize=20)
plt.xlabel("x -->",fontsize=16)
plt.ylabel("y -->",fontsize=16)
plt.legend()     #-->will print the label for all line
plt.grid(True,color='k')    #---> will show grid on graph
plt.show()
#%matplot inline -->to show line to line
```
